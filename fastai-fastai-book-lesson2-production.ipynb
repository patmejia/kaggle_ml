# Defining Search Parameters and Function for Retrieving Images of a Specific Term

```python
term = 'parsley'
min_sz=128
max_images=150 

def search_images(term:str, max_images:int, min_sz:int)->List:
    """
    This function takes a search term, a maximum number of images to retrieve, and a minimum size for the images, 
    and returns a list of image URLs that match the search term.
    """
    print(f"Searching for {term} images with a minimum sample size of {min_sz} and a maximum number of {max_images} images")
    results = ddg_images(term, max_results=max_images)
    images = [result.get('image') for result in results if result.get('width')>min_sz and result.get('height')>min_sz]
    return L(images)

parsley_images = search_images(term, max_images, min_sz)
print(parsley_images[:5]) # prints the first 5 urls of the images
```

# Title: Testing our `search_images` function


```python
def test_search_images():
    parsley_images = search_images(term, max_images, min_sz)
    # assert checks if the number of images is equal to the maximum number of images.
    assert len(parsley_images) == max_images, f'Expected {max_images} images, got {len(parsley_images)}'
    for image in parsley_images:
        # assert checks if the image starts with http.
        assert image.startswith('http'), f'Expected URL, got {image}'
    # print prints a message if all the test cases pass.
    print('All test cases pass')

test_search_images()
```
