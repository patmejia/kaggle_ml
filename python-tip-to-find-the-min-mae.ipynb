# Python tip to find the Min MAE 

An intro python tip for the Intermediate Machine Learning course. Refereces the tutorial: [**Step-1-Introduction-Exercise**](https://www.kaggle.com/code/alexisbcook/introduction#:~:text=Prerequisites-,Your,-Turn). 

It is a good practice to compare the MAE scores of different models. This tip shows how to create an array of MAE scores and find the model with the lowest MAE score.

![Ames Housing dataset image](https://i.imgur.com/lTJVG4e.png)

## **Start**

Add `mae_results = []` and `mae_results.append(mae)` to create an array that stores the mean absolute error (MAE) for each model and later serves as the input to find the minimum MAE.

Here is how:

```python
from sklearn.metrics import mean_absolute_error

# Function for comparing different models
def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)
# create a dictionary to store the results: mae_results
mae_results = []                 # <--- Add this line
for i in range(0, len(models)):
    mae = score_model(models[i])
    print("Model %d MAE: %d" % (i+1, mae))
    mae_results.append(mae)      # <--- Add this line
```

To complete **step-1** of the exercise, we find the minimum MAE using the `min()` function:

```python
# find the min of mae_results
min_mae_results = min(mae_results)

# print min mae_results
print("min mae_results: %d" % min_mae_results)

# find index of min mae_results or best model
best_model_index = mae_results.index(min_mae_results)+1

# print best model index
print("best model index: %d" % best_model_index)

# get the best model from models list
best_model = models[best_model_index-1]

print("best model: %s" % best_model)

# make best_model equals the word model_ concatenated with index of best model
best_model = "model_" + str(best_model_index)
print(best_model)
```
